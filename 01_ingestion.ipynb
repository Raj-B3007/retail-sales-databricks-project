```python

df_customers = spark.read.table("workspace.default.customers")
display(df_customers)

# df_customers.printSchema()

#Record count validation
print("Customer Count:", df_customers.count())

#Null Analysis 
from pyspark.sql.functions import col

df_customers.filter(
    col("city").isNull()
).show()

print("Null city count:",df_customers.filter(
    col("city").isNull()).count())

#duplicate check
from pyspark.sql.functions import count

(
    df_customers
    .groupBy("customer_id")
    .agg(count("*").alias("cnt"))
    .filter("cnt > 1")
    .show()
)

```
